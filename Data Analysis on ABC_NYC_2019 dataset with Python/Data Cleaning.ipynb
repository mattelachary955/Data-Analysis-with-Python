# OUTLIERS

Handling the Outliers is of types:
1. Trimming 
2. Capping

Methods of Handling the Outliers are
1. Z-score
2. IQR
3. Percentiles

```python
df6 = df3.copy()
```

```python
df6
```

```python
df6.info()
```

```python
df6.head()
```

considering "price" and "Number of reviews" for outliers

```python
plt.subplot(1,2,1)
sns.distplot(df6["price"])
plt.subplot(1,2,2)
sns.distplot(df6["number_of_reviews"])
plt.show()
```

```python
sns.boxplot(df6["price"])
```

Trimming using Interquartile Range

```python
q1 = df6["price"].quantile(0.25)
q3 = df6["price"].quantile(0.75)
iqr = q3-q1
```

```python
q1, iqr, q3
```

```python
df6.describe()
```

```python
min = q1-1.5*iqr
max = q3+1.5*iqr
```

```python
min, max
```

Trimming outliers using IQR

```python
df6[(df6["price"] < min) | (df6["price"] > max)]
```

```python
df6[(df6["price"] > min) & (df6["price"] < max)]
```

Capping outliers using IQR

```python
df6["price"] = np.where((df6["price"])> max, max, np.where(df6["price"]<min, min, df6["price"]))
```

Outlier Handling using Z-score

```python
df7 = df3.copy()
```

```python
mean = df7["price"].mean()
std = df7["price"].std()
print(mean, std)
```

```python
high = mean + 3*std
low = mean - 3*std
```

```python
high, low
```

```python
df7.isna().sum()
```

```python
df7.describe()
```

```python
df7["z-score"] = (df7["price"] - df7["price"].mean())/df7["price"].std()
```

```python
df7
```

```python
df7[(df7["z-score"]<-3) | (df7["price"] >3)]
```

Trimming outliers using Z-score

```python
df7[(df7["z-score"]>-3) & (df7["price"] <3)]
```

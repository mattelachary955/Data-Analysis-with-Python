# PREPROCESSING THE DATA

Removing the null values from dataset

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
df = data.dropna()
```

```python
df.head()
```

```python
df.columns
```

```python
average_rating = round(df["Rating"].sum()/len(df["Rating"]), 2)
```

```python
print("Average Rating of Apps-",average_rating)
```

Number of Apps with 5 Rating

```python
c1 = 0
for i in df["Rating"]:
    if i==5:
        c1+=1
print(f"There are {c1} Apps with 5 Rating")
```

Number of Apps rating between 4 and 4.5

```python
c2 = 0
for i in df["Rating"]:
    if i>4 and i<=4.5:
        c2+=1
print(f"There are {c2} Apps with 5 Rating")
```

Number of Apps rating more than average rating

```python
c3 = 0
for i in df["Rating"]:
    if i > average_rating:
        c3+=1
print(f"There are {c3} Apps with 5 Rating")
```

```python
df.head()
```

Total Number of Categories of Apps

```python
unique_categories = df["Category"].unique()
unique_categories
```

```python
print(f"There are {len(unique_categories)} category apps")
```

```python
c4 = 0
for i in df["Category"]:
    if i == "ART_AND_DESIGN":
        c4+=1
print(f"There are {c4} of 'ART_AND_DESIGN' category apps")
```

Number of Apps of all category apps

```python
category_dict = {}
for i in unique_categories:
    count = 0
    for j in df["Category"]:
        if i==j:
            count += 1
    category_dict[i] = count
    print(f"{i} - {count} apps")
category_dict
```

```python
categories = pd.DataFrame.from_dict(category_dict,orient='index')
categories
```

Number of Apps with Type

```python
df["Type"].unique()
```

```python
free = 0
for i in df["Type"]:
    if i == "Free":
        free+=1
paid = len(df["Type"])-free
print(f"There are {free} apps of 'Free' type and {paid} of 'Paid' apps")
```

# REPLACING NAN VALUES USING SCIKIT-LEARN

```python
data.head()
```

```python
data.isnull().sum()
```

```python
impute = SimpleImputer(missing_values = np.nan, strategy = "mean")
```

```python
impute.fit(data.iloc[:,2:3])
data.iloc[:,2:3] = impute.transform(data.iloc[:,2:3])
```

```python
data["Rating"]
```

```python
data["Rating"] = data["Rating"].round(1)
```

```python
data["Rating"]
```

```python
data
```

```python
data.isnull().sum()
```

```python
impute_cat = SimpleImputer(missing_values = np.nan, strategy = "most_frequent")
impute_cat.fit(data.loc[:,["Type","Content Rating", "Current Ver","Android Ver"]])
data.loc[:,["Type","Content Rating", "Current Ver","Android Ver"]] = impute_cat.transform(data.loc[:,["Type","Content Rating", "Current Ver","Android Ver"]])
```
